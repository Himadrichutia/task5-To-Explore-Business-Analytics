## Most of the items are from office supplies category and least are from technology

```python
# To check value counts of Category column
dataset["Sub-Category"].value_counts()
```

```python
# Visualising sales and profit against 
dataset.groupby('Category').agg({'Sales':'sum','Profit':'sum'})
```

```python
dataset.groupby("Category")[['Profit', 'Sales']].sum().plot.bar(color = ["blue", "green"],figsize= (8,5))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

## Here we can say that sales and profit is more in technology and less profit is in furniture

```python
# Visualising sales and profit against 
dataset.groupby('State').agg({'Sales':'sum','Profit':'sum'})
```

```python
csp = dataset.groupby("State")[["Sales", "Profit"]].sum().sort_values(by = "Sales", ascending = False)
csp[:].plot.bar(color = ["orange", "blue"],figsize= (18,8))
plt.title("Profit/Loss & Sales across States")
plt.xlabel("States")
plt.ylabel("Profit/Loss & Sales")
plt.show()
```

## Here we see  that States like California and New York have high sales with a good profit whereas staes like Texas, Pennsylvania, Ohio face huge losses

```python
# Visualising sales and profit against 
dataset.groupby('Sub-Category').agg({'Sales':'sum','Profit':'sum'})
```

```python
scsp=dataset.groupby("Sub-Category")[["Sales","Profit"]].mean().sort_values(by = "Sales", ascending = False)
scsp.plot.bar(color=['red','orange'],figsize= (18,8))
plt.title("Profit and Sales across various Sub-Categories",fontsize=20)
plt.xlabel('Sub-Category', fontsize=15)
plt.ylabel("Profit/Loss & Sales")
plt.tick_params(labelsize=14)
plt.show()
```

```python
## Sales and profit is more in sub category copies and less profit is table
```

```python
dataset.groupby("Region")[['Profit', 'Sales']].sum().plot.bar(color = ["blue", "yellow"],figsize= (8,5))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

```python
## product sub category wise profit and sales
dataset1=dataset.groupby(['Category','Sub-Category'])['Sales','Profit'].sum().reset_index()
dataset1.sort_values(by = ['Category','Sub-Category', 'Sales','Profit'],ascending=False)
```

### Visualizing segment wise sales distribution in each region

```python
plt.figure(figsize=(12,8))
plt.title("Distribution of Segment wise sales in each region " , fontsize=15)
sns.barplot(x="Region",y='Sales',data=dataset,hue='Segment',order=dataset['Region'].value_counts().index,palette='Blues')
plt.xlabel('Region', fontsize=15)
plt.tick_params(labelsize=12)
plt.show()
```
